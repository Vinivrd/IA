```python
def get_sentences(file_name):
    with open(file_name, 'r',encoding="ISO-8859-1") as f:
        return f.read().split('.')

MIN_LENGTH = 15
sentences = []
for file_name in file_names:
    sentences+=get_sentences(file_name)

sentences = [sentence.replace('\n','') for sentence in sentences]
sentences = [sentence.replace('\t','') for sentence in sentences]
sentences = [sentence for sentence in sentences if len(sentence)>MIN_LENGTH]
```

```python
pd.DataFrame(sentences)
```

```python
from collections import defaultdict
import numpy as np

def construir_matriz_transicao(sentencas):
    transicoes = defaultdict(lambda: defaultdict(int))

    for sentenca in sentencas:
        palavras = sentenca.strip().split()
        for i in range(len(palavras) - 1):
            palavra_atual = palavras[i]
            palavra_proxima = palavras[i + 1]
            transicoes[palavra_atual][palavra_proxima] += 1

    # Transformar contagens em probabilidades
    matriz_probabilidade = {}
    for palavra_atual, destinos in transicoes.items():
        total = sum(destinos.values())
        matriz_probabilidade[palavra_atual] = {palavra_proxima: contagem / total
                                               for palavra_proxima, contagem in destinos.items()}

    return matriz_probabilidade


matriz_transicao = construir_matriz_transicao(sentences)
matriz_transicao
```

```python
import random

def gerar_frase(matriz_probabilidade, comprimento_maximo=15, palavra_inicial=None):
    if not palavra_inicial:
        palavra_inicial = random.choice(list(matriz_probabilidade.keys()))

    frase = [palavra_inicial]
    palavra_atual = palavra_inicial

    for _ in range(comprimento_maximo - 1):
        if palavra_atual not in matriz_probabilidade:
            break
        proximas_palavras = list(matriz_probabilidade[palavra_atual].keys())
        probabilidades = list(matriz_probabilidade[palavra_atual].values())
        palavra_atual = random.choices(proximas_palavras, weights=probabilidades)[0]
        frase.append(palavra_atual)

    return ' '.join(frase)

frase = gerar_frase(matriz_transicao)
print(frase)
```

A distribuição estacionária te diz, no longo prazo, qual a chance de estar em cada palavra, independentemente de onde você começou.

$$
\boxed{\pi^{(k+1)} = \pi^{(k)} P}
$$


```python
import numpy as np

def calcular_distribuicao_estacionaria_iterativa(matriz_probabilidade, tol=1e-8, max_iter=5):
    palavras = list(matriz_probabilidade.keys())
    palavra_indices = {palavra: i for i, palavra in enumerate(palavras)}
    print(palavra_indices)
    n = len(palavras)
    P = np.zeros((n, n))

    # Preenche a matriz de transição
    for palavra_atual, destinos in matriz_probabilidade.items():
        i = palavra_indices[palavra_atual]
        for palavra_proxima, prob in destinos.items():
            if palavra_proxima in palavra_indices:
              j = palavra_indices[palavra_proxima]
              P[i, j] = prob

    # Vetor de probabilidade inicial (uniforme)
    pi = np.ones(n) / n

    for _ in range(max_iter):
        pi_novo = np.dot(pi, P)
        if np.linalg.norm(pi_novo - pi, 1) < tol:
            break
        pi = pi_novo

    # Normaliza
    pi = pi / pi.sum()

    return palavras, pi


palavras, distrib_estacionaria = calcular_distribuicao_estacionaria_iterativa(matriz_transicao)

for palavra, prob in zip(palavras, distrib_estacionaria):
    print(f"{palavra}: {prob:.4f}")
```

```python
def gerar_frase2(matriz_probabilidade, palavras, distrib_estacionaria, comprimento_maximo=100):
    palavra_inicial = random.choices(palavras, weights=distrib_estacionaria)[0]
    return gerar_frase(matriz_probabilidade, comprimento_maximo, palavra_inicial)


frase = gerar_frase2(matriz_transicao, palavras, distrib_estacionaria)
print(frase)
```

```python
from collections import defaultdict

def construir_matriz_transicao_pares(sentencas):
    transicoes = defaultdict(lambda: defaultdict(int))

    for sentenca in sentencas:
        palavras = sentenca.strip().split()
        if len(palavras) < 3:
            continue  # precisa de pelo menos 3 palavras para pares

        for i in range(len(palavras) - 2):
            estado_atual = (palavras[i], palavras[i + 1])
            proxima_palavra = palavras[i + 2]
            transicoes[estado_atual][proxima_palavra] += 1

    # Transformar contagens em probabilidades
    matriz_probabilidade = {}
    for estado_atual, destinos in transicoes.items():
        total = sum(destinos.values())
        matriz_probabilidade[estado_atual] = {proxima: contagem / total
                                              for proxima, contagem in destinos.items()}

    return matriz_probabilidade


matriz_pares = construir_matriz_transicao_pares(sentences)

for estado, transicoes in matriz_pares.items():
    print(f"{estado} -> {transicoes}")
```

```python
import random

def gerar_frase_pares(matriz_probabilidade, comprimento_maximo=50, estado_inicial=None):
    if not estado_inicial:
        estado_inicial = random.choice(list(matriz_probabilidade.keys()))

    frase = [estado_inicial[0], estado_inicial[1]]
    estado_atual = estado_inicial

    for _ in range(comprimento_maximo - 2):
        if estado_atual not in matriz_probabilidade:
            break
        proximas_palavras = list(matriz_probabilidade[estado_atual].keys())
        probabilidades = list(matriz_probabilidade[estado_atual].values())
        proxima_palavra = random.choices(proximas_palavras, weights=probabilidades)[0]

        frase.append(proxima_palavra)

        # Atualiza o estado: novo par (segunda palavra do par atual, nova palavra)
        estado_atual = (estado_atual[1], proxima_palavra)

    return ' '.join(frase)


# Gerar frase
frase = gerar_frase_pares(matriz_pares)
print(frase)
```

# Exemplo com PageRank

Páginas: `A, B, C, D, E, F, G, H, I, J`

**Links entre páginas:**

| Página | Links para |
|:------:|:----------:|
| A      | B, C, D |
| B      | C, E |
| C      | A, F |
| D      | E, F, G |
| E      | F |
| F      | G, H |
| G      | H |
| H      | I |
| I      | J |
| J      | A |


```python
import numpy as np

# Definindo as páginas
paginas = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
indice = {pagina: i for i, pagina in enumerate(paginas)}

# Matriz 10x10 inicializada com zeros
P = np.zeros((10, 10))

# Preenchendo a matriz de transição
P[indice['A'], [indice['B'], indice['C'], indice['D']]] = 1/3
P[indice['B'], [indice['C'], indice['E']]] = 1/2
P[indice['C'], [indice['A'], indice['F']]] = 1/2
P[indice['D'], [indice['E'], indice['F'], indice['G']]] = 1/3
P[indice['E'], [indice['F']]] = 1
P[indice['F'], [indice['G'], indice['H']]] = 1/2
P[indice['G'], [indice['H']]] = 1
P[indice['H'], [indice['I']]] = 1
P[indice['I'], [indice['J']]] = 1
P[indice['J'], [indice['A']]] = 1

P
```

```python
def calcular_pagerank_damping(P, alpha=0.85, tol=1e-8, max_iter=10):
    n = P.shape[0]
    P_prime = alpha * P + (1 - alpha) * (np.ones((n, n)) / n)

    pi = np.ones(n) / n

    for _ in range(max_iter):
        pi_novo = np.dot(pi, P_prime)
        if np.linalg.norm(pi_novo - pi, 1) < tol:
            break
        pi = pi_novo

    return pi


pagerank = calcular_pagerank_damping(P)

# Mostrar resultados
print("\nPageRank:")
for nome, valor in zip(paginas, pagerank):
    print(f"{nome}: {valor:.4f}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar o grafo dirigido
G = nx.DiGraph()

# Adicionar os nós (páginas)
G.add_nodes_from(paginas)

# Adicionar as arestas (links)
G.add_edges_from([
    ('A', 'B'), ('A', 'C'), ('A', 'D'),
    ('B', 'C'), ('B', 'E'),
    ('C', 'A'), ('C', 'F'),
    ('D', 'E'), ('D', 'F'), ('D', 'G'),
    ('E', 'F'),
    ('F', 'G'), ('F', 'H'),
    ('G', 'H'),
    ('H', 'I'),
    ('I', 'J'),
    ('J', 'A')
])



# Plotar
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_labels(G, pos)

plt.axis('off')
plt.show()
```

# Label Propagation

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random

# Carregar o Karate Club Graph
G = nx.karate_club_graph()

# Pegar os rótulos verdadeiros ('Mr. Hi' ou 'Officer')
club_map = {'Mr. Hi': 0, 'Officer': 1}

# Agrupar nós por rótulo
nodos_por_clube = {0: [], 1: []}

for node, data in G.nodes(data=True):
    label = club_map[data['club']]
    nodos_por_clube[label].append(node)

# escolher alguns nós rotulados de cada clube
random.seed(42)  # fixar aleatoriedade para reprodução
n_rotulados_por_classe = 5

rotulos_iniciais = {}
for classe, nos in nodos_por_clube.items():
    escolhidos = random.sample(nos, n_rotulados_por_classe)
    for no in escolhidos:
        rotulos_iniciais[no] = classe

print("Nós rotulados inicialmente:", rotulos_iniciais)

# Definindo as cores para cada rótulo
cores_iniciais = {0: "skyblue", 1: "lightgreen"}
node_colors_iniciais = []
for node in G.nodes():
    if node in rotulos_iniciais:
        node_colors_iniciais.append(cores_iniciais[rotulos_iniciais[node]])
    else:
        node_colors_iniciais.append("lightgray")

# Plotar o grafo
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw_networkx_nodes(G, pos, node_color=node_colors_iniciais, node_size=500)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_color="black")
plt.axis('off')
plt.show()
```

```python
def label_propagation_markov(G, rotulos_iniciais, n_classes, max_iter=100, tol=1e-8):

    n = G.number_of_nodes()

    # Matriz de transição P
    A = nx.to_numpy_array(G)
    P = A / A.sum(axis=1, keepdims=True)

    # Inicializa Y (matriz de distribuição de rótulos)
    Y = np.zeros((n, n_classes))
    for node, label in rotulos_iniciais.items():
        Y[node, label] = 1

    # print(Y)
    # print(P)

    # Propagação
    for i in range(max_iter):
        print("iteracao:",i)
        Y_novo = np.dot(P, Y)
        print(Y_novo)

        # Reinserir rótulos fixos
        for node, label in rotulos_iniciais.items():
            Y_novo[node] = np.zeros(n_classes)
            Y_novo[node, label] = 1

        if np.linalg.norm(Y_novo - Y, ord=1) < tol:
            break

        Y = Y_novo

    # Definir o rótulo final como o de maior valor
    labels_final = np.argmax(Y, axis=1)

    return labels_final


labels_final = label_propagation_markov(G, rotulos_iniciais, n_classes=2)
labels_final
```

```python
from sklearn.metrics import accuracy_score

# pegar os rótulos reais do grafo
rotulos_reais = []
mapa_rotulos = {'Mr. Hi': 0, 'Officer': 1}

for node, data in G.nodes(data=True):
    rotulo_real = mapa_rotulos[data['club']]
    rotulos_reais.append(rotulo_real)

# calcular a acurácia
acuracia = accuracy_score(rotulos_reais, labels_final)

print(f"Acurácia da Label Propagation: {acuracia * 100:.2f}%")
## eh um problema facil
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Definir as cores para as classes
cores_finais = {0: "skyblue", 1: "lightgreen"}

# Criar a lista de cores dos nós
node_colors_finais = [cores_finais[label] for label in labels_final]

# Layout do grafo
pos = nx.spring_layout(G, seed=42)

# Plotando
plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_color=node_colors_finais, node_size=500)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_color="black")
plt.title("Resultado da Label Propagation (Cadeia de Markov)", fontsize=16)
plt.axis('off')
plt.show()
```

# Cluster Label Propagation

```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def cluster_label_propagation_markov(G, max_iter=10, tol=1e-8):

    n = G.number_of_nodes()

    # Matriz de adjacência e matriz de transição
    A = nx.to_numpy_array(G)
    P = A / A.sum(axis=1, keepdims=True)

    # Inicializa Y: cada nó começa com seu próprio rótulo (grupo)
    Y = np.eye(n)

    # Propagação
    for _ in range(max_iter):
        Y_novo = np.dot(P, Y)

        if np.linalg.norm(Y_novo - Y, ord=1) < tol:
            break

        Y = Y_novo

    # Após propagação, o rótulo do nó é o índice do valor máximo
    labels_final = np.argmax(Y, axis=1)

    return labels_final
```

```python
# Carregar o grafo
G = nx.karate_club_graph()

# Rodar o cluster label propagation
labels_final = cluster_label_propagation_markov(G)
print(labels_final)

# Visualizar os clusters
n_clusters = len(set(labels_final))
cores = plt.get_cmap('tab10', n_clusters)



pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_color=[cores(label) for label in labels_final], node_size=500)
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_color="black")
plt.title("Cluster Label Propagation via Cadeias de Markov", fontsize=16)
plt.axis('off')
plt.show()
```
